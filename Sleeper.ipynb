```python
users_df_cleaned = users_df[['user_id','display_name']]
users_df_cleaned
```

```python
final_users_df = pd.merge(left=users_df_cleaned, right=rosters_df_cleaned, how='left', left_on='user_id', right_on='owner_id')
final_users_df
```
